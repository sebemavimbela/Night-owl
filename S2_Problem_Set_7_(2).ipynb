**Examples 1.)** A company wants to open a new store in one of three potential locations: Grahamstown, Port Alfred and Port Elizabeth. The company considers two main factors: average monthly sales and store rental costs. The Average Monthly Sales in Grahamstown, Port Alfred and Port Elizabeth are R 50000, R 70000 and R 60000, respectively. The Store Rental Costs in Grahamstown, Port Alfred and Port Elizabeth are R 15000, R 20000 and R 18000, respectively.

The company wants to choose the town that offers the best potential for profit by maximizing the average monthly sales relative to the rental costs. Write a Python code to use matrix operations to calculate the town that will be the best choice for the company."

**HINT:** This question involves calculating a profitability metric for each town (average monthly sales minus rental costs) and then selecting the town with the highest profitability.

```python
import numpy as np

# Data
averageSales = np.array([50000, 70000, 60000])
rentalCosts = np.array([15000, 20000, 18000])

# Calculate the profitability (average sales - rental costs) for each town
profitability = averageSales - rentalCosts # Element-wise subtraction

# Find the index of the town with the highest profitability
bestTownIndex = np.argmax(profitability)

print(profitability,'\n')
# Town names list
townNames = ['Grahamstown', 'Port Alfred', 'Port Elizabeth']

# Get the name of the best town
bestTown = townNames[bestTownIndex]

print(f"'{bestTown}' is the best location for the new store.")
```

**Example 2.)** A family is planning a vacation and considering four different destinations -  Grahams Hotel, Victoria Hotel, Contry Hotel and Joza Resort. They are evaluating these options based on three factors - travel time, accommodation cost, and available activities. It takes 5, 3, 2 and 6 hours to travel to Grahams Hotel, Victoria Hotel, Contry Hotel and Joza Resort, respectively. Accommodation cost R 300, R 150, R 200, and R 100 at Grahams Hotel, Victoria Hotel, Contry Hotel and Joza Resort, respectively. A preference score (1 to 10) of 8, 6, 9 and 4 were assiged to Grahams Hotel, Victoria Hotel, Contry Hotel and Joza Resort, respectively for the available of activities in each destination.

The family wants to maximize their experience by choosing the destination that has the shortest travel time, lowest accommodation cost, and the highest preference score for activities. Write a Python code to use matrix operations to calculate the best destination for the family.

**Weighted Average**

To generate weighted average for this problem we consider that the family wants to maximize their experience by choosing the destination that has the shortest travel time, lowest accommodation cost, and the highest preference score for activities. Since they want to minimize travel time and accommodation cost while maximizing activity preference, we can assign weights accordingly.

To achieve this, we assign positive weights to the factors you want to maximize (activity preference) and negative weights to the factors you want to minimize (travel time and accommodation cost). Hence:

 - Travel time: -1 (minimize)
 - Accommodation cost: -1 (minimize)
 - Activity preference: 1 (maximize)

```python
# Given data
travelTime = np.array([5, 3, 2, 6])
accommodationCost = np.array([300, 150, 200, 100])
activityPreference = np.array([8, 6, 9, 4])

# Weighted average weights
weights = np.array([-1, -1, 1])

# Combine travel time, accommodation cost, and activity preference into a matrix
criteriaCatrix = np.array([travelTime, accommodationCost, activityPreference])

# Calculate the weighted score for each destination
scores = np.dot(criteriaCatrix.T, weights)
# scores = (criteriaCatrix.T * weights).sum(axis=1) # This's the same as the above

# Find the index of the destination with the highest score
bestDestinationIndex = np.argmax(scores)
# bestDestinationIndex = np.where(scores==scores.max())[0][0] # Same as the above

# Destination names list
destinationNames = ['Grahams Hotel', 'Victoria Hotel', 'Contry Hotel','Joza Resort']

# Get the name of the best destination
bestDestination = destinationNames[bestDestinationIndex]

print(scores, '\n')

print(f"The best destination for the family is '{bestDestination}'.")
```

```python
# 4.) Your solution here.
import numpy as np

# Data from the table for seven suppliers and five items
item_prices = np.array([
    [10, 15, 20, 25, 30],  # Supplier 1
    [12, 18, 22, 28, 32],  # Supplier 2
    [9, 14, 19, 24, 29],   # Supplier 3
    [11, 16, 21, 26, 31],  # Supplier 4
    [13, 17, 23, 27, 33],  # Supplier 5
    [8, 13, 18, 23, 28],   # Supplier 6
    [14, 19, 24, 29, 34]   # Supplier 7
])

# Delivery costs for each supplier
delivery_costs = np.array([50, 45, 55, 48, 52, 40, 58])

# Assuming the owner wants to buy one of each item
quantities = np.array([1, 1, 1, 1, 1])

# Calculate the total cost of items for each supplier
# This is the dot product of the item_prices matrix and the quantities vector
total_item_cost = np.dot(item_prices, quantities)

# Add the delivery cost to get the final total cost for each supplier
final_total_costs = total_item_cost + delivery_costs

# Find the index of the supplier with the minimum total cost
best_supplier_index = np.argmin(final_total_costs)

# Create a list of supplier names
supplier_names = [f"Supplier {i+1}" for i in range(7)]

# Get the name of the best supplier
best_supplier = supplier_names[best_supplier_index]

print(f"The supplier offering the best bid is: {best_supplier}")
print(f"\nThe total cost for each supplier (items + delivery) is:")
print(final_total_costs)
```
